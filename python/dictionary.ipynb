```python
def word_count(text):
    words = text.split(' ')
    ans = {}

    for word in words:
        if word in ans:
            ans[word] += 1
        else:
            ans[word] = 1

    return ans
```

```python
word_count('apple orange apple banana orange')
```

```python
def find_anagrams(text) :
    words = text.split(' ')
    ans = {}

    for word in words :
        sorted_word = ''.join(sorted(word))

        if sorted_word in ans :
            ans[sorted_word].append(word)
        else :
            ans[sorted_word] = [word]
    
    return ans
```

```python
find_anagrams("listen silent enlist inlets hello world")
```

```python
expenses = {}

def update_expenses(name, amount) :
    if name in expenses:
        expenses[name] += amount
    else:
        expenses[name] = amount

def print_expenses() :
    for key, value in expenses.items():
        print(key, " : ", value)
```

```python
update_expenses("food", 300)
```

```python
print_expenses()
```
