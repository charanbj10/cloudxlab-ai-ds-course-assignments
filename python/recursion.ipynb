```python
# recursion

def factorial(n) :
    if (n==0) : return 1

    return n * factorial(n-1)
```

```python
factorial(5)
```

```python
# multiply using recursion without *

def multiply_recur(a,b, cntr) :
    if (b > a) : a,b = b,a

    if (cntr == abs(b - 0)) : return 0
    
    ans = abs(a - 0) + multiply_recur(a, b, cntr+1)

    if (((a < 0 and b > 0) or (a > 0 and b < 0)) and cntr == 0) : return 0 - ans
    return ans
```

```python
multiply_recur(-3, 7, 0)
```

```python
# Find Power

def findPower(x, n) :
    if (n == 0) : return 1
    if (n < 0) :
         return 1 / x * findPower(x, n+1)
    return x * findPower(x, n-1)
```

```python
findPower(2, -3)
```

```python
# Find Quotient & Reminder

def recursive_divide(dividend, divisor) :
    if (dividend - divisor < 0) :
        return (0, dividend)

    dividend -= divisor

    q, r = recursive_divide(dividend ,  divisor)
    return (q+1, r)
```
