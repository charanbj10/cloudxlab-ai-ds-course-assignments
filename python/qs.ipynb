```python
# We are designing thermometer by measure the electric current. We noted when current is 1amp, 
# the temperature is 5C and when current is 2amps, the temperature is 30C. Now, come up with an equation 
# which give temperature when we inoput current. T = i * m + c Basically, find out values of m and c.


# Find Equation

# y = ax + b Find a & b

def find_equation(dataset) :
    var1 = dataset[0]
    var2 = dataset[1]


    a  = (var1[1] - var2[1]) /  (var1[0] - var2[0] )
    print(a)

    b = var1[1] - (var1[0]*a)
    print(b)

    return a, b

```

```python
dataset = [
    [1, 5],
    [2, 30]
]

a, b = find_equation(dataset)
print("y = {a}x + {b}")
```
