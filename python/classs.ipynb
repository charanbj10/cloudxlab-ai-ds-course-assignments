```python
class Vector :
    def __init__(self, coordinates):
        

    def add(self, move) :
       return Vector(self.x + move.x, self.y + move.y)

    
```

```python
position = Vector(3, 4)
move = Vector(-2, 3)

new_position = position.add(move)
print(new_position.x, new_position.y)
```

```python
new_position = position.add(move) # (1, 7)
new_position.dist(position) # print 3.605551275463989 (sqrt(13))

wind1 = Vector(10, 0)
# def multiply by scaler (int, float)
move_wind1 = wind1.multiply(2) 
new_position = new_position.add(move_wind1)

wind2 = Vector(5, 3)
move_wind2 = wind2.multiply(2) 
new_position = new_position.add(move_wind2)
```
