```python
# Zoom IN :
import numpy as np
def magnify(image, factor) :

    answer = []
    row = len(image)
    col = len(image[0])

    final_rows = row + (row - 1) * factor
    final_cols = col + (col - 1) * factor

    final_arr = np.zeros((final_rows, final_cols))

    # row-wise linear interpollation
    r_indx = 0
    c_index = 0
    for i in range(0, row) :
        for j in range(0, col-1) :
            val1 = image[i][j]
            val2 = image[i][j+1]

            sum = (val2 - val1) / (factor + 1 )
            
            final_arr[r_indx][c_index] = val1
            c_index += 1

            for x in range(1, factor+1) :
                final_arr[r_indx][c_index] = val1 + (sum * x)
                c_index += 1

            final_arr[r_indx][c_index] = val2

        r_indx += (factor + 1)
        c_index = 0
    
    # col-wise linear interpollation
    r_indx = 0
    c_index = 0
    
    for j in range(0, len(final_arr[0])) :
        for i in range(0, len(final_arr) - factor - 1, factor+1) :
            
            val1 = final_arr[i][j]
            val2 = final_arr[i+factor+1][j]

            sum = (val2 - val1) / (factor + 1 )

            for x in range(1, factor+1) :
                final_arr[i+x+1][j] = val1 + (sum * x)
            
    return final_arr
    
```
