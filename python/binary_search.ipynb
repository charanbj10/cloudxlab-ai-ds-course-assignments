```python
def sqrt(left, right, target) :
    guess = (left + right)/2
    print(left, right)

    if abs(guess * guess - target) <= 0.0001:
        return guess

    if (guess * guess > target) :
        right = guess
    else :
        left = guess
    return sqrt(left, right, target)
```

```python
sqrt(0, 80, 80)
```

```python
def cube(left, right, target) :
    guess = (left + right)/2
    print(left, right)

    if abs(guess * guess * guess - target) <= 0.0001:
        return guess

    if (guess * guess * guess > target) :
        right = guess
    else :
        left = guess
    return cube(left, right, target)
```

```python
cube(0, 27, 27)
```

```python
def compute_nth_root(left, right, target, pow) :
    guess = (left + right)/2
    print(left, right)

    if abs(guess ** pow - target) <= 0.0001:
        return guess

    if (guess ** pow > target) :
        right = guess
    else :
        left = guess
    return compute_nth_root(left, right, target, pow)
```

```python
compute_nth_root(0, 81, 81, 4)
```

```python
def findLog10(left, right, target) :
    guess = (left + right) / 2

    if abs(10 ** guess - target) < 1e-7 :
        return guess
    
    if (10 ** guess > target) :
        right = guess
    else :
        left = guess
    
    return findLog10(left, right, target)
```

```python
findLog10(0,100, 100)
```

```python
def findLogN(left, right, target, pow) :
    guess = (left + right) / 2

    if abs( pow ** guess - target) < 1e-7 :
        return guess
    
    if (pow ** guess > target) :
        right = guess
    else :
        left = guess
    
    return findLogN(left, right, target, pow)
```
