```python
# Calculate Interest

def findInterest(p, r, t) :
    return (p * r * t)/100
```

```python
findInterest(1500, 4.3, 3)
```

```python
# Calculate Hypotenuse

def findHypotenuse(b, h):
    return (( b*b ) + ( h*h ))**0.5
```

```python
findHypotenuse(5,12)
```

```python
# Find Distance 2D

def find_distance_2d(x1, y1, x2, y2) :
    return ((x2 - x1)**2 + (y2 - y1)**2)**0.5
```

```python
find_distance_2d(1, 2, 4, 6)
```

```python
# Find Distance 3D

def find_distance_3d(x1, y1, z1, x2, y2, z2) :
    return ((x2 - x1)**2 + (y2 - y1)**2 + (z2 - z1)**2)**0.5
```

```python
find_distance_3d(1, 2, 3, 4, 6, 9)
```

```python
# Find Manhattan Distance

def find_manhattan(x1, y1, x2, y2) :
    return abs(x1 - x2) + abs(y1 - y2)
```

```python
find_manhattan(-1, -1, 1,1)
```

```python
# Predict Y on straight line

def predict_y(m, x, c) :
    return (m*x)+c
```

```python
predict_y(2,4,3)
```

```python
# Find Slope & Intercept

def fit(x1, y1, x2, y2) :
    m = (y2 - y1)/(x2 - x1)

    c = y1 - (m * x1)

    return (m, c)
```

```python
fit(0,5,2,9)
```

```python
# Find Derivative

def square(x):
    return x * x

def cube(x):
    return x * x * x

def find_derivative(func, x, h) :
    return (func(x+h) - func(x)) / h
```

```python
find_derivative(cube, 2, 0.0001)
```
