```python
import math

class Vector:
    
    def __init__(self, coordinates):
        self.coordinates = coordinates
    

    def add(self, vector):
        newcoordinate = []
        for index, value in enumerate (vector.coordinates) :
            newcoordinate.append(vector.coordinates[index]+self.coordinates[index])
        return tuple(newcoordinate) # Return new Vector

    def multiply(self, scalar):
        newcoordinate = []
        for index, value in enumerate (self.coordinates) :
            newcoordinate.append(scalar*self.coordinates[index])
        return tuple(newcoordinate) # Return new Vector

    def veloctity_when_turned_by_theta(self, theta):
        newcoordinate=[]
        newcoordinate.append(math.cos(math.radians(theta)))
        newcoordinate.append(math.sin(math.radians(theta)))
        return tuple(newcoordinate)

x= Vector((1,1))
y= Vector((2,2))
z= Vector(x.add(y))
print (z.coordinates)
z1= Vector(z.multiply(2))
print (z1.coordinates)
```

```python
import numpy as np
import math

def rotate(points, angle):
    theta = angle*math.pi/180
    s, c = math.sin(theta), math.cos(theta)
    # x1 = x cos - y * sin
    # y1 = x sin + y cos

    rot_mat = np.array([
        [c, s],
        [-s, c]
    ])

    return np.matmul(points, rot_mat)
```

```python
import matplotlib.pyplot as plt
import numpy as np

points = np.array([[1, 1], [3, 4]])

angle = 0
for _ in range(360):
    angle += 1
    points = rotate(points, angle)

    plt.scatter(points[:, 0], points[:, 1])
    plt.plot(points[:, 0], points[:, 1]) 

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Rotated Points")
plt.grid(True)

plt.show()

```
