```python
# Check Which Point Closer to P

def check_closer_to_p(P, A, B) :
    x1, y1 = A
    x2, y2 = B
    p1, p2 = P

    val_1 = ((x1 - p1)**2 + (y1 - p2)**2)**0.5
    val_2 = ((x2 - p1)**2 + (y2 - p2)**2)**0.5

    print(val_1, val_2)

    if val_1 < val_2 :
        return 'A'
    elif val_1 > val_2 :
        return 'B'
    else :
        return 'EQUAL'
```

```python
check_closer_to_p((4, 4), (0, 0), (8, 8))
```

```python
# Is Point P on Line in 1D?

def is_point_on_line_1d(a,b,p) :
    if (p >= a and p <= b) :
        return True
    elif (p <= a and p >= b) :
        return True
    return False
```

```python
is_point_on_line_1d(-1, 3, 0)     
```

```python
# Are Two Lines Overlapping or Touching in 1D

def are_lines_touching_or_overlapping(a,b,c,d) :
    if ((c >= a and c <= b) or (d >= a and d <= b)) :
        return True
    elif ((a >= c and b <= d) or (a >= c and b <= c)) :
        return True
    else :
        return False
```

```python
are_lines_touching_or_overlapping(-2, -7, -4, -3)
```

```python
# Is a Point Inside a Rectangle (with Sides Parallel to the Axes)?

def is_point_inside_rectangle(x1, y1, x2, y2, p1, p2) :
    if ((p1 >= x1 and p2 >= y1) and (p1 <= x2 and p2 <= y2)) :
        return True
    return False
```

```python
is_point_inside_rectangle(-5, -5, 5, 5, 0, 0)
```

```python
# Are Rectangles Intersecting?

def are_rectangles_intersecting(rect1: tuple, rect2: tuple) :
    ax1, ay1 = rect1[0]
    ax2, ay2 = rect1[1]

    bx1, by1 = rect2[0]
    bx2, by2 = rect2[1]

    if (ax2 >= bx1 and ax1 <= bx2) and (ay2 >= by1 and ay1 <= by2) : return True
    return False
    
```

```python
# Rectangles overlap partially
# are_rectangles_intersecting(((0, 0), (3, 3)), ((2, 2), (5, 5)))  
# Output: True

# # One rectangle is completely to the right of the other
# are_rectangles_intersecting(((0, 0), (1, 1)), ((2, 2), (3, 3)))  
# # Output: False

# # Touching at corner
# are_rectangles_intersecting(((0, 0), (2, 2)), ((2, 2), (4, 4)))  
# # Output: True

# # One rectangle inside another
are_rectangles_intersecting(((0, 0), (5, 5)), ((1, 1), (2, 2)))  
# # Output: True
```

```python
# Impurity formula

# impurity = (2 * min(c1, c2)) / (c1 + c2)
# own formula : 1 - abs(c1 - c2) / (c1 + c2)

def impurity (c1, c2) :
    return 1 - (abs(c1 - c2)/(c1 + c2))
```
