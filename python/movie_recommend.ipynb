```python
import pandas as pd
import numpy as np
df = pd.read_csv('movies_ratings.csv')
```

```python
df = df.drop(columns=['Harshita', 'Priya', 'Yukta'])
```

# Primary Objective

1. Find similar people -  cross recommendation - People like you have like these movies too.
2. Find similar movies - People who like this movie also like that movie. 

```python
# Take average of each row & replace it with Nan for cleaning the data

rows = df.shape[0]

for i in range(0, rows) :
    movie_df = df.iloc[i]
    # print(movie_df, "\n")
    
    sum_val= 0
    count = 0
    for j in range(1, movie_df.shape[0]) :

        if not pd.isna(movie_df.iloc[j]):
            sum_val += movie_df.iloc[j]
            count += 1

    avg_val = sum_val / count

    for j in range(1, movie_df.shape[0]) :
        if pd.isna(movie_df.iloc[j]):
            df.iloc[i, j] = avg_val
    
df
    
```

```python
# ChatGPT Code using Numpy for above -- for better performance

import numpy as np

values = df.iloc[:, 1:].to_numpy(dtype=float)

row_means = np.nanmean(values, axis=1)

# Find NaN positions and replace
inds = np.where(np.isnan(values))
values[inds] = np.take(row_means, inds[0])

# Put back into DataFrame
df.iloc[:, 1:] = values
```

```python
def min_max_scaling(arr) :
    mini = np.min(arr)
    maxi = np.max(arr)

    # print("Maximum Value = ", maxi, "\nMinimum Value = ", mini)

    arr = (arr - mini) / (maxi - mini)

    return arr
```

```python
# min_max_scaling(arr)


for col in df.columns[1:]:
    arr = np.array(df[col].to_numpy(dtype=float))
    
    df[col] = min_max_scaling(arr)

df
```
