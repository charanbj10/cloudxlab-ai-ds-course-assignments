```python
def find_min_max(arr) :
    mini = arr[0]
    maxi = arr[0]
    for i in arr :
        mini = min(i, mini)
        maxi = max(i, maxi)
    
    return (maxi, mini)

```

```python
find_min_max([5, 8, 2, 10, 3])  
# Output: (2, 10)

# find_min_max([7, 7, 7, 7])  
# # Output: (7, 7)
```

```python
def min_max_normalize(value, data) :
    mini = min(data)
    maxi = max(data)

    normalized_val = (value - mini) / (maxi - mini)

    return normalized_val
```

```python
min_max_normalize(10, [10, 20, 30])
```

```python
def compute_mean(arr) :
    length = len(arr)
    if (length == 0) : return 0

    sum = 0
    for i in arr : 
        sum += i
    
    return sum / length
```

```python
compute_mean([2, 4, 4, 4, 5, 5, 7, 9]) 
```

```python
# Compute Standard Deviation (SD) of a List of Numbers

def compute_sd(arr) :
    mean = compute_mean(arr)

    sqr_arr = []
    for i in arr:
        sqr_arr.append(abs(i - mean) ** 2)
    
    average = compute_mean(sqr_arr)
    
    return average ** 0.5
```

```python
compute_sd([2, 4, 4, 4, 5, 5, 7, 9])
```

```python
# Detect Outliers with Z-Score

def find_outliers(arr, threshold) :
    mean = compute_mean(arr)
    sd = compute_sd(arr)

    print(mean, sd)

    outlier = []
    for i in arr :
        z = (i - mean) / sd
        if (z > threshold) :
            outlier.append(i)
    
    return outlier
```

```python
find_outliers([5, 6, 7, 8, 9, 10, 100], 2)  
```

```python
import numpy as np

# Compute Interquartile Range (IQR)
def compute_iqr(arr): 
    data = np.array(arr)

    q1 = np.percentile(data, 25)
    print(q1)
    q3 = np.percentile(data, 75)
    print (q3)

    return int(q3 - q1)
```

```python

compute_iqr([1, 2, 3, 4, 5, 6, 7, 8, 9]) 
```

```python
# Standardization

def standardize(arr) :
    mean = compute_mean(arr)
    sd = compute_sd(arr)

    print(mean, sd)

    standardized_arr = []
    for i in arr :
        if (sd == 0) : standardized_arr.append(0)
        else : standardized_arr.append((i - mean) / sd)

    return standardized_arr
```

```python
standardize([1, 2, 3, 4, 5]) 
```

```python
# Compute Root Mean Square Error (RMSE) in n-dimensions

def compute_rmse(arr1, arr2) :
    diff = []
    for i in range(0, len(arr1)) :
        subtr = arr1[i] - arr2[i]
        diff.append(subtr ** 2)
    
    return compute_mean(diff)
```

```python
compute_rmse([2, 3, 4], [3, 1, 7])  
```

```python
# Compute Mean Absolute Error in N-Dimensions

def flatten(x):
    if isinstance(x, (list, tuple)):
        result = []
        for item in x:
            result.extend(flatten(item))
        return result
    return [x]

def compute_mae(actual, predicted):
    actual_flat = flatten(actual)
    predicted_flat = flatten(predicted)

    print(actual_flat, predicted_flat)

    sum = 0
    for i in range(0, len(actual_flat)) :
       sum += abs(actual_flat[i] - predicted_flat[i])
    
    return sum/len(actual_flat)
    
```

```python
# Example 1: 1-D
# actual = [3, 5, 2]
# predicted = [2, 5, 4]

# Example 2: 2-D
# actual = [[1, 2], [3, 4], [5, 6]]
# predicted = [[2, 2], [2, 5], [5, 7]]

# Example 3: 3-D
actual = [
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
]

predicted = [
    [[2, 1], [4, 3]],
    [[6, 5], [8, 9]]
]

compute_mae(actual, predicted)  
```
