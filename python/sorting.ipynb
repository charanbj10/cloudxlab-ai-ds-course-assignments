```python
def merge(arr1, arr2, ptr1, ptr2) :
    merged_arr = []
    while ( ptr1 < len(arr1) and ptr2 < len(arr2)) :
        if (arr1[ptr1] <= arr2[ptr2]) :
            merged_arr.append(arr1[ptr1])
            ptr1 += 1
        else :
            merged_arr.append(arr2[ptr2])
            ptr2 += 1
    
    while(ptr1 < len(arr1)) :
        merged_arr.append(arr1[ptr1])
        ptr1 += 1
    
    while(ptr2 < len(arr2)) :
        merged_arr.append(arr2[ptr2])
        ptr2 += 1
    
    return merged_arr


def merge_sort(arr, low, high) :
    mid = ( low + high ) // 2

    arr1 = []
    if (low == high) :
        arr1.append(arr[low])
        return arr1

    left_arr = merge_sort(arr, low, mid)
    right_arr = merge_sort(arr, mid+1, high)

    return merge(left_arr, right_arr, 0, 0)


def sort(arr) :
    print(merge_sort(arr, 0, len(arr)-1))

arr = [5,6,2,3,1,6]
sort(arr)
```
