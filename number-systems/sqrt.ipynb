```python
def sqrt(left, right, target) :
    guess = (left + right)/2
    print(left, right)

    if abs(guess * guess - target) < 1e-3:
        return guess

    if (guess * guess > target) :
        right = guess
    else :
        left = guess
    return sqrt(left, right, target)
```
