```python
base = 5
```

```python
def subtraction(a, b) :
    ans = 0
    digits = 1
    borrow = 0

    while(a > 0 or b > 0 or borrow) :
        last_a = a % 10
        last_b = b % 10

        print(last_a, last_b)
        minus = last_a - last_b - borrow

        if (minus < 0) :
            last_a += base
            minus = last_a - last_b - borrow
            borrow = 1
        else :
            borrow = 0

        print(last_a, last_b)

        ans += digits * minus
        
        a = a // 10
        b = b // 10

        digits *= 10

    print(ans)

```

```python
subtraction(101, 2)
```
