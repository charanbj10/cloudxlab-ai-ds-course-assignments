```python
def findLog10(left, right, target) :
    guess = (left + right) / 2

    if abs(10 ** guess - target) < 1e-3 :
        return guess
    
    if (10 ** guess > target) :
        right = guess
    else :
        left = guess
    
    return findLog10(left, right, target)
```

```python
findLog10(0,10,10)
```
