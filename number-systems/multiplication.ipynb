```python
base = 5
```

```python
def multiplication(a, b) :
    ans = 0
    total = 0
    carry = 0
    digits = 1
    demo_a = a
    
    def_cntr = 1

    while(b > 0 or carry) :
        last_b = b % 10
        while( a > 0 or carry) :
            last_a = a % 10

            multiply = (last_a * last_b) + carry
            carry = 0
            if (multiply >= base) :
                ans += (multiply % base) * digits
                carry = multiply // base
            else :
                ans += (multiply * digits)
            
            a = a // 10
            digits *= 10
        
        # add ans to total
        place = 1
        temp_tot = total
        total = 0
        cntr = 1
        sub_carry = 0

        print(ans, temp_tot)
        while(ans > 0 or temp_tot > 0 or sub_carry) :
            if ((cntr < def_cntr and temp_tot != 0) or (temp_tot == 0 and place == 1 and def_cntr > 1)) :
                tot_val = temp_tot % 10
                total += (place * tot_val)
                temp_tot = temp_tot // 10
            else :
                tot_val = (temp_tot % 10) + (ans % 10) + sub_carry
                sub_carry = 0

                if (tot_val >= base) :
                    total += (tot_val % base) * place
                    sub_carry = (tot_val // base)
                else :    
                    total += (place * tot_val)

                temp_tot = temp_tot // 10
                ans = ans // 10
            
            place *= 10
            cntr += 1
        
        def_cntr += 1
        
        b = b // 10
        a = demo_a
        ans = 0
        digits = 1
    
    print(total)



```
