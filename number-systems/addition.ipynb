```python
base = 5
```

```python
# works only for single & double digits answers. 
# Eg : 44 + 1 returns 50, expected answer = 100

def addition(a, b) :

    # convert to base 10
    base_a = ((a // 10) * base) + (a % 10)
    base_b = ((b // 10) * base) + (b % 10)

    print(base_a)

    # sum and convert to respective base
    added_val = base_a + base_b 

    base_val = ((added_val // base) * 10) + (added_val % base)

    print(base_val)
```

```python
# pen and paper logic

def addition(a, b) :
    ans = 0
    digits = 1
    carry = 0

    while(a > 0 or b > 0 or carry) :
        last_a = a % 10
        last_b = b % 10

        sum = last_a + last_b + carry
        carry = 0
        if (sum >= base) :
            rem = sum % base
            carry = 1
            ans += digits * rem
        else :
            ans += digits * sum
        
        digits *= 10

        a = a // 10
        b = b // 10

    print(ans)

```

```python
addition(44,1)
```

```python
# ChatGPT Code

def add_in_base(a, b, base):
    a = str(a)
    b = str(b)

    i, j = len(a) - 1, len(b) - 1
    carry = 0
    result = []

    while i >= 0 or j >= 0 or carry:
        digit_a = int(a[i]) if i >= 0 else 0
        digit_b = int(b[j]) if j >= 0 else 0

        total = digit_a + digit_b + carry
        carry = total // base
        result.append(str(total % base))

        i -= 1
        j -= 1

    return ''.join(reversed(result))
```

```python
add_in_base(44, 1, 5)
```
