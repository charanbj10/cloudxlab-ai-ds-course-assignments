```python
base = 5
```

```python
def write_to_3digit() :
    mul = 0
    digit = 1
    addr = 10
    total_loops = base * base

    result = []
    cntr = 0
    cntr_limit = 1
    for i in range(total_loops) :
        for j in range(base) :
            result.append(j + mul)
        
        cntr += 1
        if (cntr >= cntr_limit-1) :
            digit *= 10
            addr = 10
            mul = digit

            cntr = 0
            cntr_limit *= base
        else :
            if ((cntr)%base == 0) :
                addr *= 10
                mul += (addr - ((cntr-1) * (addr // 10)))
                # addr = 10
            else :
                mul += addr
        
        print(" Mul Value : ", mul, addr, cntr)

    
    return result

        
```
